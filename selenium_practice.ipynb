```python
# Importing the required libraries
import pandas as pd 
import numpy as np 
pd.plotting.register_matplotlib_converters()
#%matplotlib notebook
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity="all"
```

```python

glassdoor_data=pd.read_csv('glassdoor_data.csv', index_col=None)
glassdoor_data.shape
#glassdoor_data.head()
glassdoor_data=glassdoor_data[glassdoor_data["Salary Estimate"]!="1"]
glassdoor_data.shape
#glassdoor_data.head()
```

```python
#b=glassdoor_data[["Salary Estimate"]].split()
salary_cleaned = glassdoor_data["Salary Estimate"].apply(lambda x : x.split('(')[0].replace('CA$',' ').replace('K',' '))

glassdoor_data['salary_min']= salary_cleaned .apply(lambda x : int(x.split('-')[0]) )
print(glassdoor_data['salary_min'].dtype)
glassdoor_data['salary_max']= salary_cleaned .apply(lambda x : int(x.split('-')[1]) )
print(glassdoor_data['salary_max'].dtype)
glassdoor_data['salary_avg']=(glassdoor_data.salary_max+glassdoor_data.salary_min)/2
glassdoor_data['Company Name']= glassdoor_data["Company Name"] .apply(lambda x : x.split('\n')[0]) 
glassdoor_data.head()
```
